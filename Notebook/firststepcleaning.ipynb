Now let us take the top 15 values only from the above series as there are very less counts for the specilizations which do not come in  the top 15
Also, Let us name the specializations which are not in the top 15 as 'Others'

```python
top_15_specializations= ['electronics and communication engineering',
'computer science & engineering' ,               
'information technology',                         
'computer engineering',                           
'computer application',                           
'mechanical engineering',                         
'electronics and electrical engineering',         
'electronics & telecommunications',                
'electrical engineering',                          
'electronics & instrumentation eng',               
'instrumentation and control engineering',         
'information science engineering',                 
'electronics and instrumentation engineering',     
'civil engineering',                               
'electronics engineering',                         
'biotechnology']
```

```python
for i in range(salary_data.shape[0]):
    if salary_data.loc[i,'Specialization'] not in top_15_specializations:
        salary_data.loc[i,'Specialization'] = 'Other'
        
```

```python
salary_data['Specialization'].value_counts()
```

Let us scale some columns where the stuents could have rated themselves on a scale of 1 to 10 

```python
scale1to10columns = ['conscientiousness','agreeableness','extraversion','nueroticism','English','Logical','Quant','Domain','ComputerProgramming'] 
```

```python
for col in scale1to10columns:
    salary_data[col] = (salary_data[col] - min(salary_data[col]))*10/(max(salary_data[col])-min(salary_data[col]))
```

```python
salary_data[scale1to10columns].head()
```

The average of conscientiousness,agreeableness,extraversion,nueroticism will give us the 'personality rating' and the average of English,Logical and Quant will give us the 'Aptitude rating' 

```python
salary_data['personality score'] = (salary_data['conscientiousness']+salary_data['agreeableness']+salary_data['extraversion']+10-salary_data['nueroticism'])/4
salary_data['aptitude score'] = (salary_data['English']+salary_data['Logical']+salary_data['Quant'])/3
```
